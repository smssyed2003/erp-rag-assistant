## 3. Functions

```python
# Simple function: normalize text
def normalize_text(text):
    """Clean up text by removing extra spaces"""
    return " ".join(text.strip().split())

# Test it
messy_text = "  Hello    world   \n  test  "
clean_text = normalize_text(messy_text)
print(f"Before: '{messy_text}'")
print(f"After: '{clean_text}'")
```

```python
# Function with multiple parameters
def create_message(role, text):
    """Create a message dictionary"""
    return {
        "role": role,
        "text": text
    }

# Use it
user_msg = create_message("user", "How to create PO?")
bot_msg = create_message("bot", "To create PO: 1. Go to MM01...")

print(f"User message: {user_msg}")
print(f"Bot message: {bot_msg}")
```

```python
# Function from our actual project: loading JSON
def load_json_demo(filename):
    """
    Load JSON file safely.
    This is similar to our backend/app/utils.py
    """
    try:
        import json
        with open(filename, 'r') as f:
            data = json.load(f)
        return data
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found")
        return None

# We'll simulate this
print("(Function defined, would use: load_json_demo('data.json'))")
```

## 4. Classes and Objects

```python
# Simple class: ChatMessage
class ChatMessage:
    def __init__(self, role, text):
        self.role = role
        self.text = text
    
    def display(self):
        return f"[{self.role.upper()}]: {self.text}"

# Create objects
user_msg = ChatMessage("user", "How to create PO?")
bot_msg = ChatMessage("bot", "To create PO: 1. Go to MM01 transaction...")

print(user_msg.display())
print(bot_msg.display())
```

```python
# Class like in our project: MemoryManager
class MemoryManager:
    def __init__(self):
        self.sessions = {}  # Store conversations
    
    def add_message(self, session_id, question, answer):
        if session_id not in self.sessions:
            self.sessions[session_id] = []
        self.sessions[session_id].append({
            "q": question,
            "a": answer
        })
    
    def get_history(self, session_id):
        if session_id in self.sessions:
            history = self.sessions[session_id]
            return f"{len(history)} messages stored"
        return "No history"

# Use it
memory = MemoryManager()
memory.add_message("user1", "What is P2P?", "P2P is Procure-to-Pay...")
memory.add_message("user1", "And O2C?", "O2C is Order-to-Cash...")

print(f"User 1 history: {memory.get_history('user1')}")
print(f"User 2 history: {memory.get_history('user2')}")
```

## 7. Practical Example: Building a Simple Chat Memory

```python
class SimpleChatMemory:
    """Simplified version of our actual MemoryManager"""
    
    def __init__(self):
        self.sessions = {}
    
    def add(self, session_id, question, answer):
        """Add Q&A pair to session"""
        if session_id not in self.sessions:
            self.sessions[session_id] = []
        
        self.sessions[session_id].append({
            "question": question,
            "answer": answer
        })
    
    def get_last_n(self, session_id, n=3):
        """Get last N exchanges"""
        if session_id not in self.sessions:
            return ""
        
        history = self.sessions[session_id]
        recent = history[-n:]  # Last n
        
        text = ""
        for item in recent:
            text += f"User: {item['question']}\n"
            text += f"Assistant: {item['answer']}\n\n"
        
        return text

# Test it
chat = SimpleChatMemory()

# Simulate a conversation
chat.add("john", 
         "What is P2P?",
         "P2P (Procure-to-Pay) is the process of purchasing goods.")

chat.add("john",
         "Is it different from O2C?",
         "Yes, O2C is Order-to-Cash (selling), P2P is buying.")

chat.add("john",
         "How many steps in P2P?",
         "Usually 5 main steps: PR, PO, GR, Invoice, Payment.")

# Get history
print("Last 2 exchanges:")
print(chat.get_last_n("john", n=2))
```
